#### 1. Write a program to generate an array of 30 numbers (1 to 30) of all possible shapes.

```python
import numpy as np

def genShapes(a):
    """
        This function generates all possible of numpy array
        arg - numpy array
    """
    cnt = 1
    l = len(a)
    
    while cnt <= l:
        if l%cnt==0:
            x = int(l/cnt)
            y = cnt
            print(a.reshape(x,y))
        cnt+=1
    
a = np.array(range(1,31))
genShapes(a)
```

#### 5. Write a program to replace first 3 numbers with 100 and middle 3'numbers with 200 and last 3 numbers with 300 in output of que.5

```python
def replaceNum(a):
    """
        replaces first 3 numbers with 100 and middle 3'numbers
        with 200 and last 3 numbers with 300
        arg- numpy array
    """
    a[:3] = 100
    a[49:52] = 200
    a[-1:-4:-1] = 300
    print(a)
    
replaceNum(a)
```

#### 6. Write a program to create a 5x5 matrix of random int numbers between 10 to 100 and

- replace lower triangle with zeros and print as a new matrix

```python
a = np.random.randint(10,100,25).reshape(5,5)
print(a)
```
